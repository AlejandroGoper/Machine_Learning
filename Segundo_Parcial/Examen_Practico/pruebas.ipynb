```python
import numpy as np 
import pandas as pd
from sklearn.datasets._samples_generator import make_blobs
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix
```

```python
x,y = make_blobs(n_samples=400,centers=4,cluster_std=0.9,random_state=6)
```

```python
def graficar(x,y):
    data = pd.DataFrame(x,columns=["x","y"])
    data["clase"] = y
    sns.relplot(x="x",y="y",data=data,hue="clase")
    plt.show()
```

```python
graficar(x,y)
```

```python
from Minima_Distancia.clasificador_minima_distancia import MDC
mdc = MDC()
mdc.fit(x,y)
```

```python
x_t,y_t = make_blobs(n_samples=100,centers=4,cluster_std=0.9,random_state=6)
y_p = mdc.predict(x_t)
```

```python
graficar(x_t,y_p)
```

```python
cm = confusion_matrix(y_t,y_p)
```

```python
print(cm)
```
