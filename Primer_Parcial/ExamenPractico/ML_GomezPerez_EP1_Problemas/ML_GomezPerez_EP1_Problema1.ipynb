# Algoritmo k - Means (Problema 1)

```python
%matplotlib inline
from numpy import genfromtxt,array
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

## Importando datos

```python
#Importamos todo el dataset
# Sera una matriz de 600x2 donde las columnas representan las dimensiones X e Y de los puntos respectivamente
dataset = genfromtxt('dataset1.csv', delimiter = ',' )
```

## Graficando el dataset

```python
#El scatterplot requiere como parametros un arreglo con los puntos de la dimensin X e Y, s representa el tamao de los puntos y c el color
scatter(dataset[:, 0], dataset[:, 1], s=20,c='green');
```

# Implementando el algoritmo K-Means de la librera Sci-kit learn

Uilizaremos el algoritmo Kmeans de la libreria SciKit Learn para encontrar los centros del dataset y as, clasificarlo.

```python
#Dado que nos piden encontrar 5 grupos, inicializamos la clase KMeans con 5 grupos y creamos el objeto kmeans
kmeans = KMeans(n_clusters=5)
#Invocamos el metodo fit del objeto kmeans y pasamos como parametro el dataset
kmeans.fit(dataset)
#Para conocer los centros de cada grupo o clase invocamos el atributo de nuestro objeto kmeans siguiente:
#Esto contiene un array de k puntos en 2D
centroides = kmeans.cluster_centers_
#Para ver a que grupo o clase, pertenece cada punto en el data set, invocamos el mtodo predict de nuestro objeto kmeans que requiere como paraetros, el dataset
#Esto contiene un array de 600 numeros enteros (un numero por cada clase o grupo)
dataset_predicted_labels = kmeans.predict(dataset)
```

## Graficando resultados, sealando los centroides de cada clase en rojo

```python
# Asignamos un color a cada punto en la grafica dependiendo de su etiqueta en dataset_predicted_labels dependiendo del estilo de color Pastel2 propio de matplotlib 
#los puntos son de tamao 20
scatter(dataset[:, 0], dataset[:, 1], c=dataset_predicted_labels, s=20, cmap='Pastel2')
#Graficamos los centroides encontrados para cada clase 
scatter(centroides[:, 0], centroides[:, 1], c='firebrick', s=700,marker='2')
```
